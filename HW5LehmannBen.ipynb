```python
from bokeh.io import show, output_file, output_notebook
from bokeh.plotting import figure
import pandas as pd
```

```python
lemon = pd.read_csv("https://raw.githubusercontent.com/asukul/DS201/master/datasets/Lemonade2016-2.csv")
```

```python
lemon.iloc[8,0] = '7/8/2016'  #Locate the NA value and replace it with a Data 

leaflet_mean = round(lemon['Leaflets'].mean(),1) 
#print(leaflet_mean)
lemon.loc[20,'Leaflets'] = leaflet_mean  #Locate the NA value of Leaflets

#print(lemonade.duplicated())  #Check for duplicates, the 6th row needs to be removed

new = lemon.drop_duplicates()  #Drop duplicate
new = new.reset_index(drop=True)
```

```python
new.head(5)
```

```python
new.loc[:,'Sales']= new['Lemon'] + new['Orange']   #Create a new column for Sales and Revenue
new.loc[:,'Revenue'] = new['Sales'] * new['Price']
new
```

```python
output_file("HW5-Lehmann-plot1.html")
```

```python
output_notebook()

p = figure(title="Sales by Temp", x_axis_label="Temp",y_axis_label="Sales")
p.circle(new['Temperature'],new['Sales'], size=new['Revenue']/2,color="orange",fill_alpha=0.5)

show(p)
```

```python
output_file("HW5-Lehmann-plot2.html")
```

```python
from bokeh.layouts import gridplot
output_notebook()

s1 = figure(background_fill_color="#fafafa",x_axis_label="Leaflets",y_axis_label="Revenue")
s1.circle(new['Leaflets'],new['Revenue'],size=12, color="#F408F8",alpha=0.6)

s2 = figure(background_fill_color="#fafafa",x_axis_label="Leaflets",y_axis_label="Sales")
s2.triangle(new['Leaflets'],new['Sales'],size=12, color="#c02942", alpha=0.6)

s3 = figure(background_fill_color="#fafafa",x_axis_label="Leaflets",y_axis_label="Temp")
s3.square(new['Leaflets'],new['Temperature'],size=12,color="#d95b43",alpha=0.6)

grid = gridplot([[s1, s2,s3]], width=250, height=250)
show(grid)
```
