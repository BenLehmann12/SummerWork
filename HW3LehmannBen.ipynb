```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
videos = pd.read_csv("https://raw.githubusercontent.com/asukul/DS201/master/datasets/USvideos.csv")
videos.head(5)
```

```python
videos.shape
```

```python
videos.info()
```

```python
new = videos.sort_values('views',ascending=False).drop_duplicates('video_id').sort_index()
```

```python
new.shape
```

```python
new.nlargest(100, "views")
```

```python
new.nlargest(10, ['likes','views'])
```

```python
df_top100 = new.nlargest(100,"views")
```

```python
df_top100.shape
```

```python
sns.pairplot(new,vars=["category_id","views","likes"],hue="category_id",palette="deep") 
plt.show()
```

```python
sns.pairplot(df_top100, vars=["category_id","views","likes"],hue="category_id",palette="deep")
plt.show()
```
